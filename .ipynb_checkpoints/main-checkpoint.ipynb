## Clearing Artifacts
Artifacts are interference caused by events such as eye blinking. ICA (Independent Component Analysis) is a method used for correcting them.

```python
ica = mne.preprocessing.ICA(n_components=20, random_state=0)
ica.fit(raw.copy().filter(8, 35))
```

```python
ica.plot_components(outlines="head")
```

"Bad" components (those that store eye blinking etc.) are stored in an ICA object

- find_bads_eog finds 'bad' channels 
- 'EOG 061' is the label associated with eye-blinking artifacts
- The threshold parameter contains a maximum allowed corellation value between ICA components and EOG channels (standard deviations)

```python
channel_names = raw.info['ch_names']
print(channel_names[-1])
```

```python
bad_idx, scores = ica.find_bads_eog(raw, 'EOG 061', threshold=2)
print(bad_idx)
```

```python
raw.plot();
```

```python
ica.apply(raw.copy(), exclude=ica.exclude).plot()
```

```python
events = mne.find_events(raw)
```

I'm not really sure why but channel 014 is used by default, the following can be used to find events across all stimulus channels.

Events are stored as (time in samples, zero, trigger)

```python
import numpy as np

stim_channels = [ch_name for ch_name in raw.info['ch_names'] if 'STI' in ch_name]
all_events = []
for stim_channel in stim_channels:
    print(f"Finding events in channel: {stim_channel}")
    events = mne.find_events(raw, stim_channel=stim_channel)
    all_events.append(events)

all_events = np.vstack(all_events) if all_events else []

if all_events.size > 0:
    all_events = all_events[all_events[:, 0].argsort()]

print("All events found:")
print(all_events)
```

```python
mne.viz.plot_events(all_events[:200])
```

After finding events, simuli can be mapped to labels which are used for creating epochs:

```python
event_id = {
    'Stimulus A': 1,
    'Stimulus B': 2,
    'Stimulus C': 3,
    'Stimulus D': 4,
    'Stimulus E': 5,
    'Stimulus F': 32
}

# properly handle simultaneous events (event_repeated)
epochs = mne.Epochs(raw, all_events, event_id=event_id, event_repeated='merge', preload=True) 
epochs.plot()
```

```python
epochs = ica.apply(epochs, exclude=ica.exclude)
epochs.apply_baseline(baseline=(None, 0)) # applying baseline correction to handle DC shift
```
